```python
mpg = df["mpg"]
```

```python
mean = mpg.mean()
print("Mean =", mean)
```

```python
median = mpg.median()
print("Median =", median)
```

```python
print(df["mpg"].mode())
```

```python
variance = mpg.var()
print("Variance =", variance)
```

```python
std = mpg.std()
print("Standard Deviation =", std)
```

```python
Q1 = mpg.quantile(0.25)
print(Q1)
```

```python
Q2 = mpg.quantile(0.50)
print(Q2)
```

```python
Q3 = mpg.quantile(0.75)
print(Q3)
```

```python
IQR = Q3 - Q1
print("IQR =", IQR)
```

```python
lower = Q1 - 1.5 * IQR
```

```python
upper = Q3 + 1.5 * IQR
```

```python
outliers = df[(df["mpg"] < lower) | (df["mpg"] > upper)]

print(outliers)
```

```python
clean_df = df[(df["mpg"] >= lower) & (df["mpg"] <= upper)]
```

```python
plt.figure(figsize=(8,5))
plt.hist(mpg, bins=10)

plt.axvline(mean, color='red', label='Mean')
plt.axvline(median, color='green', label='Median')

plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.boxplot(mpg)

plt.title("Boxplot of MPG")

plt.show()
```
