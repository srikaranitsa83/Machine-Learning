```python
Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Q1 :", Q1)
print("Q3 :", Q3)
print("IQR :", IQR)
print("Lower Limit :", lower_limit)
print("Upper Limit :", upper_limit)
```

```python
outliers = df[(df["Age"] < lower_limit) | (df["Age"] > upper_limit)]

outliers
```

```python
outliers.shape
```

```python
df = df[(df["Age"] >= lower_limit) & (df["Age"] <= upper_limit)]
```

```python
df.shape
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Age"])
plt.title("Boxplot After Removing Outliers")
plt.savefig("../outputs/plots/adult1.png")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df.select_dtypes(include=["object", "string"]).columns
```

```python
le = LabelEncoder()
```

```python
df["Sex"] = le.fit_transform(df["Sex"])
```

```python
df["Income"] = le.fit_transform(df["Income"])
```

```python
df[["Sex", "Income"]].head()
```

```python
df = pd.get_dummies(
    df,
    columns=["Workclass", "Education"],
    drop_first=True
)
```

```python
df.head()
```

```python
df[["Age", "Hours_per_week"]].head()
```

```python
scaler = MinMaxScaler()

df[["Age", "Hours_per_week"]] = scaler.fit_transform(
    df[["Age", "Hours_per_week"]]
)
```

```python
df[["Age", "Hours_per_week"]].head()
```

```python
standard = StandardScaler()

df[["Capital_Gain", "Capital_Loss"]] = standard.fit_transform(
    df[["Capital_Gain", "Capital_Loss"]]
)
```

```python
df[["Capital_Gain", "Capital_Loss"]].head()
```

```python
df.groupby("Relationship")["Age"].mean()
```

```python
bins = [0, 25, 45, 65, 100]

labels = ["Young", "Adult", "Middle", "Senior"]

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels
)
```

```python
numeric_df = df.select_dtypes(include=["int64","float64"])
```

```python
numeric_df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
```

```python
reduced_data = pca.fit_transform(numeric_df)
```

```python
reduced_data.shape
```

```python
reduced_df = pd.DataFrame(
    reduced_data,
    columns=[
        "PC1",
        "PC2",
        "PC3",
        "PC4",
        "PC5"
    ]
)
```

```python
reduced_df.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    range(1,6),
    pca.explained_variance_ratio_
)

plt.xlabel("Principal Components")
plt.ylabel("Explained Variance")
plt.title("PCA Explained Variance")

plt.savefig("../outputs/plots/pca_variance.png")

plt.show()
```
